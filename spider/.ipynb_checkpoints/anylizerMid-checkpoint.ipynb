```python
import json
import pandas as pd
import pymongo
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt

myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["soccerData"]
mycol = mydb["midRawData"]
data = pd.DataFrame(list(mycol.find()))

data_new = data.drop_duplicates()
data_new['masterMidGoal'] = data_new['masterMidGoal'].astype(np.int64)
data_new['guestMidGoal'] = data_new['guestMidGoal'].astype(np.int64)
data_new = data_new.drop(['_id', 'time','place'], axis=1)
```

```python

def df2mongo(df, mycol):
    records = json.loads(df.T.to_json()).values()
    mycol.remove()
    result = mycol.insert_many(records)
    return result

def get18(master,guest):
    if master > guest:
        return 18
    if master < guest:
        return 81
    if master == guest:
        return 99
    
def getType(yapanMasterStartOdd, yapanGuestStartOdd, yapanPankouStart):
    linTypeStart = get18(yapanMasterStartOdd, yapanGuestStartOdd)
    return str(linTypeStart) + yapanPankouStart

def getUpDownGoal(masterGoal, guestGoal, masterMidGoal,guestMidGoal):
    masterGoalMid = masterGoal - masterMidGoal
    guestGoalMid = guestGoal - guestMidGoal
    return masterGoalMid - guestGoalMid

def getGoalMid(masterGoal, guestGoal, masterMidGoal,guestMidGoal):
    if masterGoal + guestGoal - masterMidGoal - guestMidGoal >= 4:
        return 4
    else:
        return masterGoal + guestGoal - masterMidGoal - guestMidGoal

def getResultMid(masterGoal, guestGoal, masterMidGoal,guestMidGoal):
    masterGoalMid = masterGoal - masterMidGoal
    guestGoalMid = guestGoal - guestMidGoal
    if masterGoalMid > guestGoalMid:
        return 3
    if masterGoalMid == guestGoalMid:
        return 1
    if masterGoalMid < guestGoalMid:
        return 0
    
def getResult(masterGoal, guestGoal):
    if masterGoal > guestGoal:
        return 3
    if masterGoal == guestGoal:
        return 1
    if masterGoal < guestGoal:
        return 0
    
def removeSub(pankou):
    pankou = pankou.replace("升","")
    pankou = pankou.replace("降","")
    return pankou.strip()


def getbili(fenzi, master, peace, guest):
    return (fenzi / (master + peace + guest)) * 100

def getbiliV2(fenzi,x):
    return (fenzi / (x['0'] + x['1']+ x['2']+ x['3']+ x['4'])) * 100
            
def getbiliV3(fenzi,x):
    if x == 0:
        return 0
    return (fenzi / x) * 100
```

```python
data_new['daxiaoPankouStart'] = data_new.apply(lambda x: removeSub(x['daxiaoPankouStart']), axis=1)
data_new['daxiaoPankou'] = data_new.apply(lambda x: removeSub(x['daxiaoPankou']), axis=1)
data_new['daxiaoPankouStartMid'] = data_new.apply(lambda x: removeSub(x['daxiaoPankouStartMid']), axis=1)
data_new['daxiaoPankouMid'] = data_new.apply(lambda x: removeSub(x['daxiaoPankouMid']), axis=1)
data_new['goalMid'] = data_new.apply(lambda x: getGoalMid(
        x['masterGoal'], x['guestGoal'], x['masterMidGoal'], x['guestMidGoal']), axis=1)

data_new['daxiaoTypeStart'] = data_new.apply(lambda x: getType(
        x['daxiaoMasterStartOdd'], x['daxiaoGuestStartOdd'], x['daxiaoPankouStart']), axis=1)
data_new['daxiaoType'] = data_new.apply(lambda x: getType(
        x['daxiaoMasterOdd'], x['daxiaoGuestOdd'], x['daxiaoPankou']), axis=1)
data_new['daxiaoTypeStartMid'] = data_new.apply(lambda x: getType(
        x['daxiaoMasterStartOddMid'], x['daxiaoGuestStartOddMid'], x['daxiaoPankouStartMid']), axis=1)
data_new['daxiaoTypeMid'] = data_new.apply(lambda x: getType(
        x['daxiaoMasterOddMid'], x['daxiaoGuestOddMid'], x['daxiaoPankouMid']), axis=1)

data_new['result'] = data_new.apply(lambda x: getResult(
        x['masterGoal'], x['guestGoal']), axis=1)
data_new['resultMid'] = data_new.apply(lambda x: getResultMid(
        x['masterGoal'], x['guestGoal'], x['masterMidGoal'], x['guestMidGoal']), axis=1)


data_new['goalMid'] = data_new.apply(lambda x: getGoalMid(
        x['masterGoal'], x['guestGoal'], x['masterMidGoal'], x['guestMidGoal']), axis=1)
```

```python
def staticDaxiao(data_new):
    data_new = data_new[data_new['goalMid'] >= 0]

    res = data_new[['daxiaoType','daxiaoTypeStartMid'
                    , "daxiaoTypeMid", "masterMidGoal", "guestMidGoal", "goalMid", "guestGoal"]]

    res = pd.pivot_table(data_new, index=['daxiaoType','daxiaoTypeStartMid','daxiaoTypeMid'
                                          , "masterMidGoal", "guestMidGoal"], columns=["goalMid"], aggfunc="count")

    res = res["guestGoal"]

    res = res.reset_index()

    a1 = res.fillna(0.0)

    a1.columns = ['daxiaoType','daxiaoTypeStartMid'
                    , "daxiaoTypeMid","masterMidGoal","guestMidGoal" ,'0', '1', '2','3', '4']


    a1['4Per'] = a1.apply(lambda x: getbiliV2(
        x["4"],x), axis=1)
    a1['3Per'] = a1.apply(lambda x: getbiliV2(
        x["3"],x), axis=1)
    a1['2Per'] = a1.apply(lambda x: getbiliV2(
        x["2"],x), axis=1)
    a1['1Per'] = a1.apply(lambda x: getbiliV2(
        x["1"], x), axis=1)
    a1['0Per'] = a1.apply(lambda x: getbiliV2(
        x["0"], x), axis=1)

    a1['da2Per'] = a1['4Per'] + a1['3Per'] + a1['2Per']
    
    a1['xi2Per'] = a1['1Per'] + a1['0Per']
    
    
    a1['k1Per'] = a1['4Per'] + a1['3Per'] + a1['2Per'] + a1['1Per']
    
    a1['k2Per'] = a1.apply(lambda x: getbiliV3(
        (x["2"] + x['3'] + x['4']), (x['1'] + x["2"] + x['3'] + x['4'])), axis=1)
    
    a1['k3Per'] = a1.apply(lambda x: getbiliV3(
       (x['3'] + x['4']), (x["2"] + x['3'] + x['4'])), axis=1)
    
    a1['k4Per'] = a1.apply(lambda x: getbiliV3(
       (x['4']), (x['3'] + x['4'])), axis=1)
    
    a1 = a1[(a1['0'] + a1['1'] + a1['2'] + a1['3'] + a1['4']) > 10]
    
    
#     a75 = a1[(a1['4Per'] > 65) | (a1['3Per'] > 65) | (a1['2Per'] > 65) | (
#         a1['2Per'] > 65) | (a1['da2Per'] > 65) | (a1['xi2Per'] > 65)]
    
    return  a1.drop(columns=['3Per', '2Per', '4Per', '1Per', '3Per', '0Per', '3Per', '3Per'])
```

```python
a75 = staticDaxiao(data_new)
a75.sort_values(by = 'da2Per',ascending=False).head(20)
```
