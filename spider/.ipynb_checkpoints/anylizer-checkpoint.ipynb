```python
import json
import pandas as pd
import pymongo
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt

myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["soccerData"]
mycol = mydb["newRawData"]
data = pd.DataFrame(list(mycol.find({"time":{"$lt":"2020-01-01 00:00:00"}})))
data = data[['masterGoal', 'guestGoal', 'startMasterOdd', 'startPeaceOdd',
             'startGuestOdd', 'startReturnRate', 'masterOdd', 'peaceOdd', 'guestOdd', 'returnRate']]
data_new = data.drop_duplicates()
data_new.info()
```

```python
def getResult(masterGoal, guestGoal):
    if masterGoal > guestGoal:
        return 3
    elif masterGoal == guestGoal:
        return 1
    else:
        return 0


def getUpDown(startOdd, endOdd):
    if endOdd < startOdd:
        return 3
    elif endOdd == startOdd:
        return 1
    else:
        return 0


def getAllFlow(master, peace, guest):
    try:
        print
        sum = str(master).split('.')[
            0]+str(peace).split('.')[0]+str(guest).split('.')[0]
        return sum
    except:
        print("error")


def simpleAllOdds(master, peace, guest):
    return int(str(master).split('.')[0]+str(peace).split('.')[0]+str(guest).split('.')[0])


def simpleOdds(odd):
    return int(str(odd).split('.')[0])


def getOddType(master, peace, guest):
    flag = 1
    oddList = [master, peace, guest]
    oddDict = {master: 1, peace: 1, guest: 1}
    sortList = sorted(oddList)
    for index in range(0, len(sortList) - 1):
        if sortList[index] != sortList[index+1]:
            oddDict[sortList[index]] = flag
            flag += 1
            if index == 1:
                oddDict[sortList[index+1]] = flag
        else:
            if index == 1:
                oddDict[sortList[index+1]] = flag
    result = ""
    for key in oddList:
        result += str(oddDict[key])
    return (int(result))

def getbili(fenzi, master, peace, guest):
    return (fenzi / (master + peace + guest)) * 100

def df2mongo(df, mycol):
    records = json.loads(df.T.to_json()).values()
    mycol.remove()
    result = mycol.insert_many(records)
    return result

def FlowChange(data_new):
    data_new['result'] = data_new.apply(
        lambda x: getResult(x['masterGoal'], x['guestGoal']), axis=1)
    data_new['masterFlow'] = data_new.apply(
        lambda x: getUpDown(x['startMasterOdd'], x['masterOdd']), axis=1)
    data_new['peaceFlow'] = data_new.apply(
        lambda x: getUpDown(x['startPeaceOdd'], x['peaceOdd']), axis=1)
    data_new['guestFlow'] = data_new.apply(
        lambda x: getUpDown(x['startGuestOdd'], x['guestOdd']), axis=1)
    data_new['returnFlow'] = data_new.apply(lambda x: getUpDown(
        x['startReturnRate'], x['returnRate']), axis=1)
    data_new['allFlow'] = data_new.apply(lambda x: getAllFlow(
        x['masterFlow'], x['peaceFlow'], x['guestFlow']), axis=1)


    data_new['simpleStartOdd'] = data_new.apply(lambda x: simpleAllOdds(
        x['startMasterOdd'], x['startPeaceOdd'], x['startGuestOdd']), axis=1)
    data_new['simpleOdd'] = data_new.apply(lambda x: simpleAllOdds(
        x['masterOdd'], x['peaceOdd'], x['guestOdd']), axis=1)


    data_new['simpleStartMaster'] = data_new.apply(
        lambda x: simpleOdds(x['startMasterOdd']), axis=1)
    data_new['simpleStartPeace'] = data_new.apply(
        lambda x: simpleOdds(x['startPeaceOdd']), axis=1)
    data_new['simpleStartGuest'] = data_new.apply(
        lambda x: simpleOdds(x['startGuestOdd']), axis=1)
    data_new['simpleMaster'] = data_new.apply(
        lambda x: simpleOdds(x['masterOdd']), axis=1)
    data_new['simplePeace'] = data_new.apply(
        lambda x: simpleOdds(x['peaceOdd']), axis=1)
    data_new['simpleGuest'] = data_new.apply(
        lambda x: simpleOdds(x['guestOdd']), axis=1)

    data_new['startOddType'] = data_new.apply(lambda x: getOddType(
        x['startMasterOdd'], x['startPeaceOdd'], x['startGuestOdd']), axis=1)
    data_new['OddType'] = data_new.apply(lambda x: getOddType(
        x['masterOdd'], x['peaceOdd'], x['guestOdd']), axis=1)
    return data_new
```

```python
def staticFlow(data_new):
    data_new = data_new[['allFlow', 'startOddType', 'OddType', "result",
                       "simpleStartOdd", "simpleOdd", "returnFlow", "startGuestOdd"]]

    res = pd.pivot_table(data_new, index=['allFlow', 'startOddType', 'OddType',
                                        'simpleStartOdd', 'simpleOdd', "returnFlow"], columns=["result"], aggfunc="count")

    res = res["startGuestOdd"]

    res = res.reset_index()

    res.columns = ['allFlow', 'startOddType', 'OddType',
                   'simpleStartOdd', 'simpleOdd', 'returnFlow', '0', '1', '3']

    a1 = res.fillna(0.0)

    a1.sort_values(by="3", ascending=False).head(10)

    a1[['0', '3']] = a1[['3', '0']]
    a1.columns = ['allFlow', 'startOddType', 'OddType',
                  'simpleStartOdd', 'simpleOdd', 'returnFlow', '3', '1', '0']

    a1['masterPer'] = a1.apply(lambda x: getbili(
        x["3"], x['3'], x['1'], x['0']), axis=1)
    a1['peacePer'] = a1.apply(lambda x: getbili(
        x["1"], x['3'], x['1'], x['0']), axis=1)
    a1['guestPer'] = a1.apply(lambda x: getbili(
        x["0"], x['3'], x['1'], x['0']), axis=1)
    a1['31Per'] = a1['masterPer'] + a1['peacePer']
    a1['10Per'] = a1['peacePer'] + a1['guestPer']
    a1['nopeacePer'] = a1['masterPer'] + a1["guestPer"]

    a1 = a1[(a1["3"] + a1["1"] + a1["0"]) > 150]

    a75 = a1[(a1['masterPer'] > 65) | (a1['peacePer'] > 65) | (a1['guestPer'] > 65) | (
        a1['31Per'] > 70) | (a1['10Per'] > 70) | (a1['nopeacePer'] > 70)]
    return a75
```

```python
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["soccerData"]
mycol = mydb["staticFlow"]
data_new = FlowChange(data_new)
a75 = staticFlow(data_new)
df2mongo(a75, mycol)
model = a75
```

```python
def getkelly(row, staticMap, choose):
    b = -1
    p = 1
    key3 = str(getattr(row, 'allFlow')) + str(getattr(row, 'startOddType')) + str(getattr(row, 'OddType')) + \
               str(getattr(row, 'simpleStartOdd')) + str(getattr(row,
                   'simpleOdd')) + str(getattr(row, 'returnFlow')) + "3"
    key1 = str(getattr(row, 'allFlow')) + str(getattr(row, 'startOddType')) + str(getattr(row, 'OddType')) + \
               str(getattr(row, 'simpleStartOdd')) + str(getattr(row,
                   'simpleOdd')) + str(getattr(row, 'returnFlow')) + "1"
    key0 = str(getattr(row, 'allFlow')) + str(getattr(row, 'startOddType')) + str(getattr(row, 'OddType')) + \
               str(getattr(row, 'simpleStartOdd')) + str(getattr(row,
                   'simpleOdd')) + str(getattr(row, 'returnFlow')) + "0"
    if choose == 3:
        b = getattr(row, 'masterOdd') - 1
        p = staticMap.get(key3, 0)
    
    if choose == 1:
        b = getattr(row, 'peaceOdd') - 1
        p = staticMap.get(key1, 0)
            
    if choose == 0:
        b = getattr(row, 'guestOdd') - 1
        p = staticMap.get(key0, 0)

    return (b * (p/100) - (100-p)/100) / b , p


def getMax(x, y, z):
    index = 0
    maxVal = max(x, y, z)
    if maxVal == x:
        index=3
    if maxVal == y:
        index=1
    if maxVal == z:
        index=0
    return maxVal, index


def getOddIndex(result):
    if result == 3:
        return "masterOdd"
    if result == 1:
        return "peaceOdd"
    if result == 0:
        return "guestOdd"
```
